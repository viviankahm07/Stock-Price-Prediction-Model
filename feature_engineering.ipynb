```python
import pandas as pd
import os
```

```python
df = pd.read_csv('../data/prices_JNJ.csv')
df = df.drop('Date', axis = 'columns')
```

```python
df.head()
```

```python
df['ret_1d'] = df['Adj Close'].pct_change()
df['ret_5d'] = df['Adj Close'].pct_change(5)
df['red_20d'] = df['Adj Close'].pct_change(20)
df['vol_5d'] = df['ret_1d'].rolling(5).std()
df['vol_20d'] = df['ret_1d'].rolling(20).std()
df['ma_10'] = df['Adj Close'].rolling(10).mean()
df['ma_20'] = df['Adj Close'].rolling(20).mean()
df['price_to_ma20'] = df['Adj Close'] / df['ma_20'] - 1
df['vol_chg'] = df['Volume'].pct_change()
df['vol_z20'] = (df['Volume'] - df['Volume'].rolling(20).mean())/df['Volume'].rolling(20).std()
df['y_reg'] = df['Adj Close'].pct_change().shift(-1)   
df['y_class'] = (df['y_reg'] > 0).astype(int)
```

```python
df = df.dropna()
df.head()
```

```python
os.makedirs("../data/raw", exist_ok=True)
df.to_csv("../data/features_JNJ.csv")
```
